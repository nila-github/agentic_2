```python
from dataclasses import dataclass

@dataclass
class Person:
    name: str
    age: int
    city: str

    
```

```python
person = Person(name="Alice", age=30, city="New York")
print(person)
```

```python
person = Person(name="Alice", age=30, city=35)
print(person)
```

```python
## Pydantic
from pydantic import BaseModel

class Person(BaseModel):
    name: str
    age: int
    city: str
person = Person(name="Alice", age=30, city="New York")
print(person)
```

```python
person = Person(name="Alice", age=30, city="35")
print(person)
```

```python
try:
    person = Person(name="Alice", age=30, city="35")
except Exception as e:
    print("Error:", e)
print(person)
```
